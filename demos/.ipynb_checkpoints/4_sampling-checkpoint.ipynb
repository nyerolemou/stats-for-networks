# Sampling 

It is not always possible to work with the entire network, due to computational constrains. Often, we might not be able to even observe the entire network in the first place. 

1. If we want to draw conclusions about the entire network, then the solution is to take a suitable sub-network, where by suitable we mean that this sub-network is a good representation of our original network.
2. If we are trying to observe some obscure behaviour, it is not necessarily best to take a random sample of vertices (because we are unlikely to capture enough such vertices).

In the notebook we present two types of random sampling of networks:
1. Induced subgraph sampling, and
2. Snowball sampling,

which aim to solve problems (1) and (2) above.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Sampling Schemes

### Induced Subgraph Sampling

The scheme is as follows: take a random sample of vertices and take the induced subgraph on those nodes. (The *induced subgraph* contains all edges from the original graph that were present between this subset of vertices.) 

### Snowball Sampling

Pick a vertex $v$ and take its $k$-hop neighbourhood, that is, all vertices and edges at distance $k$ from $v$.

These two schemes produce very different results, which we can see in an example below.

### Example

Let's load in a dataset from the [SNAP](https://snap.stanford.edu/data/). This is the email network used in a previous notebook, with the vertices coloured by the department to which each individual belongs (42 in total).

```python
path = '../data/email-Eu-core.txt'
labelPath = '../data/email-Eu-core-department-labels.txt'
edges = np.loadtxt(path,dtype=int)
labels = list(np.loadtxt(labelPath,usecols=1,dtype=int))
G = nx.Graph()
G.add_edges_from(edges)
print('G has {} vertices and {} edges.'.format(G.number_of_nodes(),G.number_of_edges()))
```

```python
plt.figure(figsize=(10,6))
nx.draw(G, node_size=40,node_color=[labels[i] for i in G.nodes()],cmap='nipy_spectral',vmin=1,vmax=42)
```

```python
def shortest_path_distmat(G):
    '''
    Return the distance matrix of shortest paths between nodes in a graph G.
    
    Parameters: nx.Graph() G
    
    Output: nxn np.array distance matrix where n=G.number_of_nodes()
    '''
    distDict = dict(nx.shortest_path_length(G))
    N = G.number_of_nodes()
    distmat = np.full((N, N), np.inf) #default to infty if there is no path between two nodes
    nodes = [v for v in distDict.keys()]
    for n in nodes:
        paths = distDict[n]
        pathkeys = [v for v in paths.keys()]
        for k in pathkeys:
            distmat[n,k] = paths[k]
    return distmat
sampled_nodes = random.sample(G.nodes, 50)
H = G.subgraph(sampled_nodes)
snowballStart = random.randint(0,1075)
shortestPath = shortest_path_distmat(G)[snowballStart]
indices = np.where(shortestPath < 2)
snowballNodes = list(indices[0])
```

On the left, an indiced subgraph on $50$ vertices chosen uniformly at random. On the right, a snowball sample of a random vertex in a $2$-step neighbourhood. We can see that the two forms of sampling produce very different results; the induced subgraph sample is disconnected, sparse and has a fairly even distribution of colours (=research groups), whereas the snowball sample is connected, dense and has picked up a lot of vertices from one research group in particular.

```python
K = G.subgraph(snowballNodes)
subgraphs = [H,K]
fig, axes = plt.subplots(1,2,figsize=(15,10))
for k in range(2):
    ax = plt.subplot(1,2,k+1)
    ax.set_aspect('equal')
    graph = subgraphs[k]
    nx.draw_spring(graph,node_size=50,node_color=[labels[i] for i in graph.nodes()],cmap='nipy_spectral',vmin=1,vmax=42) 
```

## The Horvitz-Thomson Estimator

What conclusions can we draw about the network from the sample? Here, we present one estimator, the **Horvitz-Thomson estimator**.

### The Estimator

Suppose we have a population $U=\{1,...,N\}$ of $N$ units, each with probability $\pi_i$ of being chosen for the sample (this is one advantage of this estimator because we do not have to assume that each node, for example, has an equal chance of being selected). For each unit $i\in U$ we make an observation $y_i$ - our value of interest. For example, it could be the number of friends individual $i$ has in a social network. Let $$\tau=\sum_{i\in U}y_i$$ be the population total and $$\mu=\frac{\tau}{N}$$ be the population average. Let $$\sigma^2=\frac{1}{N}\sum_{i\in U}(y_i-\mu)^2$$ denote the population variance.

**Goal:** to estimate $\tau$ and $\mu$ from a sample $S=\{i_1,...,i_n\}\subset U$, where we have the observed values $y_i$ for all $i\in S$.

The **Horvitz-Thomson estimate** of $\tau$ is $$\hat{\tau}_\pi = \sum_{i\in S}\frac{y_i}{\pi_i}.$$ The estimate of the mean is $$\hat{\mu}_\pi = \frac{\hat{\tau}_\pi}{N}.$$ $\hat{\tau}_\pi$ is an unbiased estimator for $\tau$.

### Applying it to Networks

We are assuming our data is represented by a network so our underlying population is $$U=V\cup \{(u,v)|u,v\in V\},$$ the set of vertices along with the set of all *potential edges*. So $$N=|V|+|V|(|V|-1).$$

Let's go into more detail with induced subgraph sampling. Suppose we are sampling $n$ vertices without replacement. Then, $$\pi_i=\frac{n}{|V|}$$ for all $i\in S$, and the probability that both $i$ and $j$ (hence the edge $(i,j)$ is observed) is $$\pi_{i,j} = \frac{n(n-1)}{|V|(|V|-1)}.$$

### Example

We can estimate the number of edges using a sample of size $n$, in a simple network $G$. If we set $y_i=0$ for all $i\in V$ and $y_{i,j}=\mathbb{1}((u,v)\in E)$, then the number of edges is half of this population total $\tau$ (because each edge is counted twice).

Then, the Horvitz-Thomson estimator $\hat{\tau}_\pi$ is $$\hat{\tau}_\pi = \frac{|V|(|V|-1)}{n(n-1)}\sum_{u\in S}\sum_{v\in S}\mathbb{1}((u,v)\in E).$$
